```python
import numpy as np
import math

def zero_handler(x):
    if not x:
        return 0.000001
    return x

def arms_ab_to_ratio(a,b,round_to):
    c = math.sqrt(a**2+b**2)
    dct = {"c": round(c,round_to),
        "sin":round(a/zero_handler(c),round_to),
        "cos":round(b/zero_handler(c),round_to),
        "tan":round(a/zero_handler(b),round_to),
        "sec":round(1/zero_handler(b/zero_handler(c)),round_to),
        "csc":round(1/zero_handler(a/zero_handler(c)),round_to),
        "cot":round(1/zero_handler(a/zero_handler(b)),round_to)}
    for k,v in dct.items():
        print(k,v)
```

```python
arms_ab_to_ratio(6,10,1)
```

```python
def arms_ac_to_ratio(a,c,round_to):
    b = math.sqrt(c**2-a**2)
    dct = {"b": round(b,round_to),
        "sin":round(a/zero_handler(c),round_to),
        "cos":round(b/zero_handler(c),round_to),
        "tan":round(a/zero_handler(b),round_to),
        "sec":round(1/zero_handler(b/zero_handler(c)),round_to),
        "csc":round(1/zero_handler(a/zero_handler(c)),round_to),
        "cot":round(1/zero_handler(a/zero_handler(b)),round_to)}
    for k,v in dct.items():
        print(k,v)
```

```python
arms_ac_to_ratio(6,10,2)
```

```python
def arms_a_degree_A_to_arm_degree(a,A,round_to):
    c = a/zero_handler(np.sin(A))
    b = math.sqrt(c**2-a**2)
    dct = {"b": round(b,round_to),
           "c": round(c,round_to),
        "sin":round(a/zero_handler(c),round_to),
        "cos":round(b/zero_handler(c),round_to),
        "tan":round(a/zero_handler(b),round_to),
        "sec":round(1/zero_handler(b/zero_handler(c)),round_to),
        "csc":round(1/zero_handler(a/zero_handler(c)),round_to),
        "cot":round(1/zero_handler(a/zero_handler(b)),round_to)}
    for k,v in dct.items():
        print(k,v)
```

```python
arms_a_degree_A_to_arm_degree(7,15,2)
```
