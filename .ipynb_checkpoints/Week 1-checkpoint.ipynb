```python
def monty(swaps):
    hand=['','Q',''] #cannot use string instead like hand=' Q ' b/c immutable.
    for s in swaps:
        if s == 'L':
            hand[0],hand[1] = hand[1],hand[0]
            print(hand)
        elif s == '0':
            hand[0],hand[2] = hand[2],hand[0]
            print(hand)
        elif s == 'R':
            hand[1],hand[2] = hand[2],hand[1]
            print(hand)
    if hand.index('Q') == 0:
        return "left"
    if hand.index('Q') == 1:
        return "middle"
    if hand.index('Q') == 2:
        return "right"
    return 'others'
```

```python
name = 'IndianPythonista'
```

```python
# without using defaultdict
d = {}
for c in name:
    if c not in d:
        d[c] = 1
    else:
        d[c] += 1
print(d)
```

```python
type(d)
```

```python
# using defaultdict
from collections import defaultdict

d = defaultdict(int)
for c in name:
    d[c] += 1
    
print(d)
```

```python
type(d)
```

```python
# more advanced defaultdict
from collections import defaultdict

d = defaultdict(lambda: 100)
for c in name:
    d[c] += 1
    
print(d)
```

### Raw strings


The use of the backslash as the escape character means, that if you want to represent a backslash, you have to write `\\`.  This can get very annoying if you are writing strings that need a lot of backslashes, but not any other escaped characters. To help with this, Python offers **raw strings**: If the opening quote is preceded by an `r`, Python will not escape any characters in it.

```python
s1 = '\t'
s2 = r'\t'
print(s1, len(s1))
print(s2, len(s2))
```

```python
'\\t' == r'\t'
```

You can think of the name assignment operation as adding a name tag to an object. For example, `c = b` should be read as "assign the name `c` to the object referenced by `b`." An object can have multiple name tags. The image below visually depicts what occurred in the code demonstrated earlier.

```python
def profit(num_cupcakes, num_tubes_frosting, tubes_per_cupcake):
    max_frosted_cakes_from_tubes = num_tubes_frosting // tubes_per_cupcake
    max_frosted_cakes = min(max_frosted_cakes_from_tubes,num_cupcakes)
    leftover_frosting = num_tubes_frosting - max_frosted_cakes*tubes_per_cupcake
    leftover_cakes = num_cupcakes - max_frosted_cakes
    profit = max_frosted_cakes * 10 + leftover_frosting * 1 + leftover_cakes * 4
    return profit
```

```python
profit(10, 5, 2)
```
