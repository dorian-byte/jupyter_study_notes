# `c.Build a multilinear regression model (you can use RFE or other methods. Make sure your assessment of the model quality is based on a test performance measure such as test MSE or test RSS)`

```python
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE
from pandas.plotting import scatter_matrix
from pandas import set_option
from pandas import read_csv
```

```python
def display_names(name_arr, ranking_arr):
    res = []
    for i in range(len(ranking_arr)):
        if ranking_arr[i] == 1:
            res.append(name_arr[i])
    return res

def find_best_feature_nums(total_feature_count, x, y):
    model = LinearRegression()
    res = []
    for i in range(total_feature_count, 0, -1):
        rfe = RFE(model, n_features_to_select=i)
        fit = rfe.fit(x, y)
        score = rfe.score(x, y)
        ranking = fit.ranking_
        feature_names = ",".join(display_names(column_names, ranking))
        res.append([i, score, feature_names])
    return pd.DataFrame(res, columns=["feature nums", "score", "feature names"])

pd.set_option('display.max_colwidth', None)
print("\n" + "-" * 100 + "\n" + " " * 40 + "\033[1m RFE on Original Data \033[0m" + " " * 60 + "\n" + "-" * 100)
find_best_feature_nums(len(column_names)-1, X, y)
```

```python
print("\n" + "-" * 100 + "\n" + " " * 40 + "\033[1m RFE on Standardized Data \033[0m" + " " * 60 + "\n" + "-" * 100)
find_best_feature_nums(len(column_names) - 1, X_std, y)
```

```python
print("\n" + "-" * 100 + "\n" + " " * 40 + "\033[1m RFE on MinMax-Normalized Data \033[0m" + " " * 60 + "\n" + "-" * 100)
find_best_feature_nums(len(column_names) - 1, X_minmax, y)
```
